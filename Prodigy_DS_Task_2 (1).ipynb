```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Iris.csv")
df.head()
```

**Exploratory Data Analysis(EDA)**



```python
df.info()
```

```python
print("The no of rows in the dataset is:" ,df.shape[0])
print("The no of columns in the dataset is:",df.shape[1])
```

**Data Cleaning**

```python
print(df.isnull().sum())
```

**Performing Correlation in the dataset:**

```python
df_numerical=df.select_dtypes(np.number)
df_numerical
```

```python
correlation_matrix = df_numerical.corr()
correlation_matrix
```

**Data Visualization**

```python
# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df, hue='Species', diag_kind='kde')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for feature in df.columns[:-1]:  # Loop through all columns except the last one (assuming the last one is the target)
    sns.boxplot(x='Species', y=feature, data=df)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
sns.countplot(x='Species', data=df)  # Change 'target' to 'Species'
plt.title('Distribution of Target Classes')
plt.show()
```
